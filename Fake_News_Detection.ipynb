```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Chota Dataset: Headline aur Label (1=Real, 0=Fake)
data = {
    'Headline': [
        'NASA discovers new planet',
        'Alien found in New Delhi subway',
        'Government announces new education policy',
        'Drink lemon juice to fly like a bird',
        'Scientists find cure for common cold',
        'Moon is made of green cheese'
    ],
    'Label': [1, 0, 1, 0, 1, 0]
}
df = pd.DataFrame(data)
print("Text Data Taiyar Hai!")
```

```python
# Text ko Numbers mein badalna
cv = CountVectorizer()
X = cv.fit_transform(df['Headline'])
y = df['Label']

# Model ko sikhana
model = MultinomialNB()
model.fit(X, y)
print("Machine ne headlines analyze karna seekh liya!")
```

```python
# Nayi headline check karte hain
test_news = ["Aliens seen in India"]
test_data = cv.transform(test_news)
prediction = model.predict(test_data)

if prediction[0] == 1:
    print(f"Result for '{test_news[0]}': REAL NEWS")
else:
    print(f"Result for '{test_news[0]}': FAKE NEWS (Caution!)")
```
