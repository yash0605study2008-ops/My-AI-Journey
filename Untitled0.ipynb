```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing
```

```python
# Data load karna
housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df['Price'] = housing.target

# Hum sirf 'AveRooms' (kamre) ka use karenge price predict karne ke liye
X = df[['AveRooms']]
y = df['Price']
```

```python
# Model banana
model = LinearRegression()
model.fit(X, y) # Machine pattern seekh rahi hai
```

```python
# Graph banana
plt.scatter(X[:100], y[:100], color='blue') # Sirf 100 points dikhane ke liye
plt.plot(X[:100], model.predict(X[:100]), color='red', linewidth=3)
plt.xlabel('Average Rooms')
plt.ylabel('House Price')
plt.title('My First AI Project: House Price Prediction')
plt.show()
```
