```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

# 1. Data banana (Salary Lakhs mein, Satisfaction 1-10, Left: 1=Haan, 0=Nahi)
# Features: [Salary, Satisfaction]
X = [[5, 3], [15, 9], [4, 2], [18, 8], [6, 5], [3, 1], [20, 10], [7, 6]]
y = [1, 0, 1, 0, 0, 1, 0, 0] # 1 matlab naukri chhod di, 0 matlab nahi chhodi

# 2. Model ko sikhana (Training)
model = LogisticRegression()
model.fit(X, y)

# 3. Prediction check karna
# Maan lo kisi ki Salary 4 Lakh hai aur Satisfaction 2 hai, kya wo naukri chhodega?
test_employee = [[4, 2]]
result = model.predict(test_employee)

if result[0] == 1:
    print("AI Prediction: Employee naukri CHHOD dega (Left)")
else:
    print("AI Prediction: Employee naukri NAHI chhodega (Stay)")

# 4. Accuracy check
print("Model Accuracy:", model.score(X, y) * 100, "%")
```
